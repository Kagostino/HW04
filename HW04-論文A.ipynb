```python
#########Get weekly and monthly return##############
import os
import pandas as pd
import math

for file in os.listdir("C:\\Users\\Mark\\Desktop\\517"):
    week = [" "," "," "," "]
    month = [" "," "," "," "," "," "," "," "," "," "," "," "," "," "," "," "," "," "," "]
    file_name = "C:\\Users\\Mark\\Desktop\\517\\" + file
    df = pd.read_csv(file_name)
    close = df.iloc[:,5]
    for i in range(close.size):
        if i>=4:
            week.append((close.iloc[i]-close.iloc[i-4])/close.iloc[i-4])
        if i>=19:
            month.append((close.iloc[i]-close.iloc[i-19])/close.iloc[i-19])
    df['week_return'] = week
    df['month_return'] = month
    df.to_csv("C:\\Users\\Mark\\Desktop\\Out\\New_"+file, columns=['week_return','month_return']) 
```

```python
#################Calculate ASKSR with weekly return####################
fileW = []
valueW = []
for file in os.listdir("C:\\Users\\Mark\\Desktop\\Out"):
    try:
        df = pd.read_csv("C:\\Users\\Mark\\Desktop\\Out\\" + file)
        Sw = df['week_return'].convert_objects(convert_numeric=True).skew(skipna = True)
        Kw = df['week_return'].convert_objects(convert_numeric=True).kurtosis(skipna = True)
        Mw = df['week_return'].convert_objects(convert_numeric=True).mean(skipna = True)
        Vw = df['week_return'].convert_objects(convert_numeric=True).var(skipna = True)
        if 3*Kw-5*math.pow(Sw,2)-9<=0:
            continue
        Aw = 3*math.sqrt(3*Kw-4*math.pow(Sw,2)-9)/(Vw*(3*Kw-5*math.pow(Sw,2)-9))
        Bw = 3*Sw/(math.sqrt(Vw)*(3*Kw-5*math.pow(Sw,2)-9))
        nw = Mw-3*Sw*math.sqrt(Vw)/(3*Kw-4*math.pow(Sw,2)-9)
        Deltaw = 3*math.sqrt(Vw)*math.sqrt(3*Kw-5*math.pow(Sw,2)-9)/(Vw*(3*Kw-4*math.pow(Sw,2)-9))
        Phiw = math.sqrt(Aw*Aw-Bw*Bw)
        rfw = math.pow((1+0.02822), 1/52)-1
        Lambdaw = 1
        Astarw = 1/Lambdaw*((Bw+Aw*(nw-rfw))/math.sqrt(Deltaw*Deltaw+math.pow(nw-rfw,2)))
        ASKSR = math.sqrt(2*(Lambdaw*Astarw*(nw-rfw)-Deltaw*(Phiw-math.sqrt(Aw*Aw-math.pow(Bw-Lambdaw*Astarw,2)))))
        fileW.append(file[4:-4])
        valueW.append(ASKSR)
    except:
        continue
    
df1 = pd.DataFrame()
df1['ETF'] = fileW
df1['value'] = valueW
df1 = df1.sort_values(by=['value'],ascending=False)
df1 = df1.set_index([pd.Series([i for i in range(1,len(df1['value'])+1)])])
df1.index.name="Ranking"
print("Rank the ETFs by ASKSR that satisfy the condition stated in thesis Eq.34 with weekly returns:")
df1
```

```python
#################Calculate ASKSR with monthly return####################
fileM = []
valueM = []
for file in os.listdir("C:\\Users\\Mark\\Desktop\\Out"):
    try:
        df = pd.read_csv("C:\\Users\\Mark\\Desktop\\Out\\" + file)
        if len(df)<=20:
            continue
        Sm = df['month_return'].convert_objects(convert_numeric=True).skew(skipna = True)
        Km = df['month_return'].convert_objects(convert_numeric=True).kurtosis(skipna = True)
        Mm = df['month_return'].convert_objects(convert_numeric=True).mean(skipna = True)
        Vm = df['month_return'].convert_objects(convert_numeric=True).var(skipna = True)
        if 3*Km-5*math.pow(Sm,2)-9<=0:
            continue
        Am = 3*math.sqrt(3*Km-4*math.pow(Sm,2)-9)/(Vm*(3*Km-5*math.pow(Sm,2)-9))
        Bm = 3*Sm/(math.sqrt(Vm)*(3*Km-5*math.pow(Sm,2)-9))
        nm = Mm-3*Sm*math.sqrt(Vm)/(3*Km-4*math.pow(Sm,2)-9)
        Deltam = 3*math.sqrt(Vm)*math.sqrt(3*Km-5*math.pow(Sm,2)-9)/(Vm*(3*Km-4*math.pow(Sm,2)-9))
        Phim = math.sqrt(Am*Am-Bm*Bm)
        rfm = math.pow((1+0.02822), 1/12)-1
        Lambdam = 1
        Astarm = 1/Lambdaw*((Bm+Am*(nm-rfm))/math.sqrt(Deltam*Deltam+math.pow(nm-rfm,2)))
        ASKSR = math.sqrt(2*(Lambdam*Astarm*(nm-rfm)-Deltam*(Phim-math.sqrt(Am*Am-math.pow(Bm-Lambdam*Astarm,2)))))
        fileM.append(file[4:-4])
        valueM.append(ASKSR)
    except:
        continue

print("Rank the ETFs by ASKSR that satisfy the condition stated in thesis Eq.34 with monthly returns:")
df2 = pd.DataFrame()
df2['ETF'] = fileM
df2['value'] = valueM
df2 = df2.sort_values(by=['value'],ascending=False)
df2 = df2.set_index([pd.Series([i for i in range(1,len(df2['value'])+1)])])
df2.index.name="Ranking"
df2
```

```python
#################Calculate ASSR with weekly return, assuming investor with CARA utility####################
fileW = []
valueW = []
for file in os.listdir("C:\\Users\\Mark\\Desktop\\Out"):
    try:
        df = pd.read_csv("C:\\Users\\Mark\\Desktop\\Out\\" + file)
        Sw = df['week_return'].convert_objects(convert_numeric=True).skew(skipna = True)
        Mw = df['week_return'].convert_objects(convert_numeric=True).mean(skipna = True)
        Vw = df['week_return'].convert_objects(convert_numeric=True).var(skipna = True)
        rfw = math.pow((1+0.02822), 1/52)-1
        t = 1
        SR = (Mw-rfw)/math.sqrt(Vw)*math.sqrt(1)
        b3 = 1  #CARA utility
        ASSR = SR*math.sqrt(1+b3*Sw/3*SR)
        fileW.append(file[4:-4])
        valueW.append(ASSR)
    except:
        continue
    
df1 = pd.DataFrame()
df1['ETF'] = fileW
df1['value'] = valueW
df1 = df1.sort_values(by=['value'],ascending=False)
df1 = df1.set_index([pd.Series([i for i in range(1,len(df1['value'])+1)])])
df1.index.name="Ranking"
print("Rank the ETFs by ASSR with weekly returns:")
df1
```
